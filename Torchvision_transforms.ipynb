```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt

from PIL import Image
import numpy as np

import torch
from torchvision import transforms
```

```python
image = Image.open("Lenna.png")
```

```python
type(image)
```

```python
np_array = np.array(image)
```

```python
plt.imshow(np_array)
```

## torchvision.transforms.Scale()

Resize the input PIL Image to the given size.

```python
scale = transforms.Scale(size=(256,256))
```

```python
result = scale(image)
```

```python
result.size
```

```python
np_array = np.array(result)
```

```python
plt.imshow(np_array)
```

## torchvision.transforms.CenterCrop()

Crops the given PIL Image at the center.

```python
centerCrop = transforms.CenterCrop(size=(256,256))
```

```python
result = centerCrop(image)
```

```python
result.size
```

```python
np_array = np.array(result)
```

```python
plt.imshow(np_array)
```

## torchvision.transforms.RandomCrop()

Crop the given PIL Image at a random location.

```python
randomCrop = transforms.RandomCrop(size=(256,256))
```

```python
for i in range(3):
    result = randomCrop(image)
    np_array = np.array(result)
    plt.subplot(131+i)
    plt.imshow(np_array)
plt.show()
```

## torchvision.transforms.RandomHorizontalFlip()

Horizontally flip the given PIL Image randomly with a probability of 0.5.

```python
randomHorizontalFlip = transforms.RandomHorizontalFlip()
```

```python
for i in range(3):
    result = randomHorizontalFlip(image)
    np_array = np.array(result)
    plt.subplot(131+i)
    plt.imshow(np_array)
plt.show()
```

## torchvision.transforms.RandomSizedCrop()

Crop the given PIL Image to random size and aspect ratio.

```python
randomSizedCrop = transforms.RandomSizedCrop(size=256)
```

```python
for i in range(3):
    result = randomSizedCrop(image)
    np_array = np.array(result)
    plt.subplot(131+i)
    plt.imshow(np_array)
plt.show()
```

## torchvision.transforms.Pad()

Pad the given PIL Image on all sides with the given pad value.

```python
pad = transforms.Pad(padding=256)
```

```python
result = pad(image)
```

```python
np_array = np.array(result)
```

```python
plt.imshow(np_array)
```
